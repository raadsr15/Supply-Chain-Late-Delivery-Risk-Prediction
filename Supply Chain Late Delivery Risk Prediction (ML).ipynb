```python
# ============================================================
# Supply Chain Late Delivery Risk Prediction (Polished Master Notebook)
# Dataset: DataCoSupplyChainDataset.csv (DataCo Smart Supply Chain)
# Goal: Predict Late Delivery (binary classification)
#
# What this script includes (clean + CV-ready):
# 1) Data loading + cleaning + consistent column names
# 2) Quick EDA plots (sales distribution, customer segment, market, delivery status, shipping vs status)
# 3) Leakage-safe target creation (late_delivery) and feature prep
# 4) Proper ML preprocessing (Ordinal encoding + scaling) using Pipeline
# 5) Multi-model benchmarking (RF, KNN, GB, ExtraTrees, DT, AdaBoost, XGBoost, LightGBM*)
# 6) Confusion matrices + classification reports + CV accuracy + feature importance (tree models)
#
# Notes:
# - Uses OrdinalEncoder to keep feature names (good for feature importance readability).
# - XGBoost/LightGBM run only if installed; otherwise they are skipped gracefully.
# ============================================================

import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split, StratifiedKFold, cross_val_score
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OrdinalEncoder, StandardScaler
from sklearn.impute import SimpleImputer

from sklearn.metrics import confusion_matrix, accuracy_score, classification_report

from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier, ExtraTreesClassifier, GradientBoostingClassifier
from sklearn.neighbors import KNeighborsClassifier

# Optional models (skip if not installed)
try:
    from xgboost import XGBClassifier
    _HAS_XGB = True
except Exception:
    _HAS_XGB = False

try:
    from lightgbm import LGBMClassifier
    _HAS_LGBM = True
except Exception:
    _HAS_LGBM = False

# -----------------------------
# 0) USER CONFIG
# -----------------------------
DATA_PATH = r"F:\RL\SC-DATA\DataCoSupplyChainDataset.csv"
ENCODING = "ISO-8859-1"
RANDOM_STATE = 42
TEST_SIZE = 0.30

sns.set(style="whitegrid")

# -----------------------------
# 1) Load Data
# -----------------------------
if not os.path.exists(DATA_PATH):
    print(f"[!] File not found at: {DATA_PATH}")
    print("    Update DATA_PATH to the correct location of DataCoSupplyChainDataset.csv")
    raise FileNotFoundError(DATA_PATH)

df = pd.read_csv(DATA_PATH, encoding=ENCODING)
print("Raw shape:", df.shape)
print(df.head(3))

# -----------------------------
# 2) Basic Cleaning / PII removal (keeps your original intent)
# -----------------------------
# Fill missing zipcodes with 0 (as in your code)
if "Customer Zipcode" in df.columns:
    df["Customer Zipcode"] = df["Customer Zipcode"].fillna(0)
if "Order Zipcode" in df.columns:
    df["Order Zipcode"] = df["Order Zipcode"].fillna(0)

# Fill missing first name
if "Customer Fname" in df.columns:
    df["Customer Fname"] = df["Customer Fname"].fillna("NotDetermined")

# Create Full Name (optional; WARNING: typically not a great ML feature due to leakage/noise)
if {"Customer Fname", "Customer Lname"}.issubset(df.columns):
    df["Customer Full Name"] = df["Customer Fname"].astype(str) + " " + df["Customer Lname"].astype(str)

# Drop columns that are sensitive or mostly useless for prediction (as in your code)
cols_to_drop = [
    "Customer Email", "Product Status", "Customer Password", "Customer Street",
    "Customer Fname", "Customer Lname", "Product Description", "Product Image",
    "Order Zipcode", "Customer Zipcode"
]
dfData = df.drop([c for c in cols_to_drop if c in df.columns], axis=1)
print("After drop shape:", dfData.shape)

# Clean column names
dfData.columns = [col.lower().replace(" ", "_") for col in dfData.columns]
dfData.rename(columns=lambda x: x.replace("(", "").replace(")", ""), inplace=True)

# -----------------------------
# 3) EDA (keeps your plots, cleaner + not duplicated)
# -----------------------------
def show_eda_plots(data: pd.DataFrame):
    # Sales distribution
    if "sales" in data.columns:
        plt.figure(figsize=(10, 6))
        sns.histplot(data=data, x="sales", bins=30, kde=True)
        plt.title("Distribution of Sales")
        plt.xlabel("Sales")
        plt.ylabel("Frequency")
        plt.tight_layout()
        plt.show()

    # Customer segment distribution
    if "customer_segment" in data.columns:
        seg_counts = data["customer_segment"].value_counts()
        plt.figure(figsize=(10, 6))
        sns.barplot(x=seg_counts.index, y=seg_counts.values)
        plt.title("Customer Segment Distribution")
        plt.xlabel("Customer Segment")
        plt.ylabel("Count")
        plt.xticks(rotation=45)
        plt.tight_layout()
        plt.show()

        plt.figure(figsize=(8, 8))
        plt.pie(seg_counts.values, labels=seg_counts.index, autopct="%1.1f%%", startangle=140)
        plt.title("Distribution of Customer Segments")
        plt.axis("equal")
        plt.tight_layout()
        plt.show()

    # Market distribution
    if "market" in data.columns:
        market_counts = data["market"].value_counts()
        plt.figure(figsize=(12, 6))
        sns.barplot(x=market_counts.index, y=market_counts.values)
        plt.title("Market Distribution")
        plt.xlabel("Market")
        plt.ylabel("Count")
        plt.xticks(rotation=45)
        plt.tight_layout()
        plt.show()

    # Delivery status distribution
    if "delivery_status" in data.columns:
        plt.figure(figsize=(10, 6))
        order = data["delivery_status"].value_counts().index
        sns.countplot(data=data, x="delivery_status", order=order)
        plt.title("Delivery Status Distribution")
        plt.xlabel("Delivery Status")
        plt.ylabel("Count")
        plt.xticks(rotation=30)
        plt.tight_layout()
        plt.show()

    # Shipping mode vs delivery status
    if {"shipping_mode", "delivery_status"}.issubset(data.columns):
        cross_tab = pd.crosstab(data["shipping_mode"], data["delivery_status"])
        cross_tab_percent = cross_tab.div(cross_tab.sum(1), axis=0) * 100

        plt.figure(figsize=(12, 8))
        cross_tab_percent.plot(kind="bar", stacked=True, colormap="Set2")
        plt.title("Delivery Status by Shipping Mode (Percentage)")
        plt.xlabel("Shipping Mode")
        plt.ylabel("Percentage")
        plt.xticks(rotation=30)
        plt.tight_layout()
        plt.show()

# Run EDA (comment out if you want faster runs)
show_eda_plots(dfData)

# -----------------------------
# 4) Target Creation (Late Delivery = 1)
# -----------------------------
if "delivery_status" not in dfData.columns:
    raise ValueError("Expected column 'delivery_status' not found after cleaning.")

train_Data = dfData.copy()

# Binary target
train_Data["late_delivery"] = np.where(train_Data["delivery_status"].astype(str).str.lower().str.contains("late"), 1, 0)

# IMPORTANT:
# Drop columns that directly reveal delivery outcome / leakage:
# - delivery_status (used to create target)
# - late_delivery_risk (if present, it's basically the label already)
leak_cols = [c for c in ["delivery_status", "late_delivery_risk"] if c in train_Data.columns]
X = train_Data.drop(columns=leak_cols + ["late_delivery"], errors="ignore")
y = train_Data["late_delivery"].astype(int)

print("\nTarget distribution:")
print(y.value_counts(normalize=True).rename("ratio"))

# -----------------------------
# 5) Train/Test Split (stratified)
# -----------------------------
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=TEST_SIZE, random_state=RANDOM_STATE, stratify=y
)

# -----------------------------
# 6) Preprocessing Pipeline
# -----------------------------
# Drop raw date columns if present (commonly cause issues unless engineered)
# (You were dropping shipping_date_dateorders, order_date_dateorders)
date_like = [c for c in X_train.columns if "date" in c and X_train[c].dtype == object]
drop_dates = [c for c in ["shipping_date_dateorders", "order_date_dateorders"] if c in X_train.columns]
drop_cols = list(set(drop_dates + date_like))

if drop_cols:
    X_train = X_train.drop(columns=drop_cols)
    X_test = X_test.drop(columns=drop_cols)

# Identify column types
cat_cols = X_train.select_dtypes(include=["object"]).columns.tolist()
num_cols = [c for c in X_train.columns if c not in cat_cols]

numeric_pipe = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="median")),
    ("scaler", StandardScaler())
])

categorical_pipe = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="most_frequent")),
    # Ordinal encoding keeps feature names stable; good for feature importance readability
    ("encoder", OrdinalEncoder(handle_unknown="use_encoded_value", unknown_value=-1))
])

preprocess = ColumnTransformer(
    transformers=[
        ("num", numeric_pipe, num_cols),
        ("cat", categorical_pipe, cat_cols),
    ],
    remainder="drop"
)

# -----------------------------
# 7) Multi-Model Evaluation: Benchmarking Template (implemented properly)
# -----------------------------
cv = StratifiedKFold(n_splits=10, shuffle=True, random_state=RANDOM_STATE)

models = {
    "Random Forest": RandomForestClassifier(n_estimators=300, random_state=RANDOM_STATE, n_jobs=-1),
    "KNN": KNeighborsClassifier(n_neighbors=15),
    "Gradient Boost": GradientBoostingClassifier(random_state=RANDOM_STATE),
    "Extra Trees": ExtraTreesClassifier(n_estimators=500, random_state=RANDOM_STATE, n_jobs=-1),
    "Decision Tree": DecisionTreeClassifier(random_state=RANDOM_STATE),
    "Adaboost": AdaBoostClassifier(random_state=RANDOM_STATE),
}

if _HAS_XGB:
    models["XGBoost"] = XGBClassifier(
        n_estimators=400,
        learning_rate=0.05,
        max_depth=6,
        subsample=0.9,
        colsample_bytree=0.9,
        eval_metric="logloss",
        random_state=RANDOM_STATE,
        n_jobs=-1
    )
else:
    print("\n[!] XGBoost not installed -> skipping XGBoost.")

if _HAS_LGBM:
    models["LightGBM"] = LGBMClassifier(
        n_estimators=600,
        learning_rate=0.05,
        random_state=RANDOM_STATE,
        n_jobs=-1,
        verbose=-1
    )
else:
    print("[!] LightGBM not installed -> skipping LightGBM.")

# Track best models
best_test_acc = -1
best_cv_acc = -1
best_test_model_name = None
best_cv_model_name = None

results = []

# Human-friendly labels for confusion matrix
class_names = ["On-time", "Late"]

for name, model in models.items():
    # Full pipeline: preprocess -> model
    pipe = Pipeline(steps=[
        ("prep", preprocess),
        ("model", model)
    ])

    # Train
    pipe.fit(X_train, y_train)

    # Predict
    y_pred = pipe.predict(X_test)

    # Metrics
    cm = confusion_matrix(y_test, y_pred)
    test_acc = accuracy_score(y_test, y_pred)

    cv_scores = cross_val_score(pipe, X_train, y_train, cv=cv, scoring="accuracy", n_jobs=-1)
    cv_mean = cv_scores.mean()
    cv_std = cv_scores.std()

    # Track best
    if test_acc > best_test_acc:
        best_test_acc = test_acc
        best_test_model_name = name

    if cv_mean > best_cv_acc:
        best_cv_acc = cv_mean
        best_cv_model_name = name

    # Print report
    print(f"\n{name}")
    print("-" * 60)
    print("Classification Report:")
    print(classification_report(y_test, y_pred, target_names=class_names))
    print("Confusion Matrix:")
    print(cm)
    print(f"Test Set Accuracy: {test_acc*100:.2f}%")
    print(f"10-Fold CV Accuracy: {cv_mean*100:.2f}%  ( {cv_std*100:.2f}%)")

    # Confusion matrix visualization
    plt.figure(figsize=(6, 4))
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", cbar=False,
                xticklabels=class_names, yticklabels=class_names)
    plt.title(f"Confusion Matrix - {name}")
    plt.xlabel("Predicted")
    plt.ylabel("Actual")
    plt.tight_layout()
    plt.show()

    # Feature importance (tree-based models only)
    # Because we use OrdinalEncoder (no feature explosion), importance aligns to original columns.
    fitted_model = pipe.named_steps["model"]
    if hasattr(fitted_model, "feature_importances_"):
        importances = fitted_model.feature_importances_
        # ColumnTransformer outputs: num_cols then cat_cols in our setup
        feature_names = num_cols + cat_cols
        fi = pd.DataFrame({"feature": feature_names, "importance": importances}).sort_values("importance", ascending=False)

        print("\nTop Feature Importances:")
        print(fi.head(20).to_string(index=False))

        plt.figure(figsize=(10, 6))
        sns.barplot(x="importance", y="feature", data=fi.head(15))
        plt.title(f"Top 15 Feature Importances - {name}")
        plt.xlabel("Importance")
        plt.ylabel("Feature")
        plt.tight_layout()
        plt.show()

    results.append({
        "model": name,
        "test_accuracy": test_acc,
        "cv_mean_accuracy": cv_mean,
        "cv_std": cv_std
    })

# -----------------------------
# 8) Summary Table (clean, CV-friendly)
# -----------------------------
results_df = pd.DataFrame(results).sort_values(by="cv_mean_accuracy", ascending=False).reset_index(drop=True)

print("\n" + "="*70)
print("Benchmark Summary (sorted by CV mean accuracy)")
print("="*70)
print(results_df.to_string(index=False, formatters={
    "test_accuracy": lambda x: f"{x*100:.2f}%",
    "cv_mean_accuracy": lambda x: f"{x*100:.2f}%",
    "cv_std": lambda x: f"{x*100:.2f}%"
}))
print("\nBest by TEST accuracy :", best_test_model_name, f"({best_test_acc*100:.2f}%)")
print("Best by CV accuracy   :", best_cv_model_name, f"({best_cv_acc*100:.2f}%)")

# -----------------------------
# 9) Optional: Save best model pipeline
# -----------------------------
# This is a nice production touch for internships.
# Uncomment to save.
#
# import joblib
# best_name = best_cv_model_name
# best_pipe = Pipeline(steps=[("prep", preprocess), ("model", models[best_name])])
# best_pipe.fit(X_train, y_train)
# joblib.dump(best_pipe, f"best_model_{best_name.replace(' ', '_').lower()}.joblib")
# print("Saved best model pipeline.")
```
